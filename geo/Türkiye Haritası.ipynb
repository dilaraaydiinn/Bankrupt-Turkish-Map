```python
import pandas as pd
konkordato = pd.read_csv(r"C:\Users\User\Downloads\geo\geo\Konkordato.csv", encoding='latin-1',sep=';')
konkordato.head(10)
```

```python
turkiye_iller.replace("K. Maras","Kahramanmaras",inplace=True)
turkiye_iller.replace('Gümüshane', 'Gumushane',inplace=True)
turkiye_iller.replace('Kinkkale', 'Kirikkale',inplace=True)
turkiye_iller.replace('Zinguldak', 'Zonguldak',inplace=True)
turkiye_iller.replace('Afyon', 'Afyonkarahisar',inplace=True)
turkiye_iller.replace('Kütahya', 'Kutahya',inplace=True)
turkiye_iller.replace('Elazığ','Elazig',inplace = True)
turkiye_iller.replace('Bingöl','Bingol',inplace = True)
turkiye_iller.replace('Bartın','Bartin',inplace = True)
turkiye_iller.replace('Karabük','Karabuk',inplace=True)
turkiye_iller.replace('Düzce','Duzce',inplace = True)
```

```python
for i in turkiye_iller["NAME_1"].items():
    print(i)
```

```python
turkiye_iller.rename(columns = {'NAME_1':'il'},inplace = True)
konkordato.rename(columns = {"Il":"il"},inplace= True)
```

```python
turkiye_iller.head(5)
```

```python
konkordato.head(5)
```

```python
harita_data = turkiye_iller.merge(konkordato,on="il")
harita_data   
```

```python
harita_data.plot(figsize=(18,18))
```

```python
harita_data.plot(column="ilan",cmap="Blues",figsize=(15,18))
```

```python
harita_data_points = harita_data.copy()
harita_data_points["center"] = harita_data_points["geometry"].centroid
harita_data_points.set_geometry("center", inplace = True)
```

```python
harita_data_points
```

```python
import adjustText as aT
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
colors = ['#C1F1FF', '#42A4D1', '#3891BA', '#2F7FA2', '#256C8A', '#1B5973', '#00547B']  # Define a range of blue colors
n_bins = [3]  # Discretizes the interpolation into bins
cmap_name = 'custom_blues'
custom_cmap = LinearSegmentedColormap.from_list(cmap_name, colors, N=256)
```

```python
ax = harita_data.plot(figsize = (40, 12), edgecolor ="grey", column = "ilan", cmap=custom_cmap)
texts = []

for x, y, label in zip(harita_data_points.center.x, harita_data_points.center.y, harita_data_points["il"]):
    texts.append(plt.text(x, y, label, fontsize = 8))
for x, y, label in zip(harita_data_points.center.x, harita_data_points.center.y, harita_data_points["ilan"]):
    texts.append(plt.text(x, y, label, fontsize = 8))

aT.adjust_text(texts, force_points=0.3, force_text=(5, 5), expand_points=(1,1), expand_text=(1,1))


ax.set_title("Türkiye'de İllere Göre Konkordato İlan Dağılımı",size = 20)
ax.set_axis_off()
```
